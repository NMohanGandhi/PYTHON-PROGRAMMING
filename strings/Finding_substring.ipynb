# String methods finding substring

# find

```python
# Returns index of first occurrence of the given substring. 
#If it is not available then we will get -1
```

```python
gmail = "mohanyadav8374@gmail.com"
```

```python
print(gmail.find('@'))
```

```python
print(gmail.find('.'))
```

```python
print(gmail.find('z'))
```

```python
print(gmail.find('gmail'))
```

# index with boundary

```python
print(gmail.find('@',7,15))   # beginingIndex,endIndex-1  : 15-1=14
```

```python
print(gmail.find('.',20,21))
```

```python
print(gmail.find('z',7,15))
```

```python
print(gmail.find('gmail',0))
```

```python
print(gmail.find('.',20))
```

 # rfind

```python
# right to left searching
```

```python
print(gmail.rfind('.'))
```

```python
# same as find but it searches string right to left(BACKWARD DIRECTION)(ending to begining)
```

```python
# here also we are using index with boundary [startingIndex,endingIndex-1]
```

```python
print(gmail.rfind('.',20))
```
