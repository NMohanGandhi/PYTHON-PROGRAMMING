```python
s1 = input("Enter first string")
s2 = input("Enter second string")
```

```python
if s1==s2:
    print("Both are same strings")
else:
    print("Both are different")
```

```python
if s1.lower()==s2.lower():
    print("Both are same strings")
else:
    print("Both are different")
```
